Artificial Intelligence for Female in Science, Technology, Engineering and Mathematics (STEM) (AI4FS) aims at establishing a hub for females in Artificial Intelligence (AI) in Nigeria in relation to Sustainable Development Goals (SDG) 4, 5 and 10 in hub-spoke universities collaboration model. The project will also be addressing the skills gaps and barriers to industries associated with female gender in STEM related fields.

The STEM-Up with AI Skill workshop is part of the AI4FS program and is being anchored by the Artificial Intelligent Systems Research Group (ARISG) residing in the Department of Mathematical Sciences, Olabisi Onabanjo University (OOU), Ago-Iwoye, Nigeria.

The goal of this workshop is to provide STEM students and faculties with exposure to data science concepts and skills that will help them integrate into society by allowing them to engage in fields outside of the traditional classroom of computer science curriculum. The demand for technical research skills led to the creation of this workshop, hands-on training that will include python for data science, excel, business intelligence,  and writing up research output for Artificial Intelligence (AI) and Machine Learning (ML).

Core constructs for the participants at the end of the program are:
Confidence: Self-efficacy to understand, contribute and succeed in data analysis, business intelligence, machine learning modelling
Community: Sense of belonging among peers, mentors and faculty in STEM for AI
Skills: Actionable knowledge to conduct, publish, present and pursue AI research professionally
Motivation:  Intention to pursue graduate school and research careers in computing and allies
Intersectional capital: Power, advocacy and access to navigate systemic barriers by validating studentsâ€™ intersectional identities.

## lambda expressions

```python
def times2(var):
    return var*2
```

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)
mytripler = myfunc(3)

print(mydoubler(11))
print(mytripler(11))
```

## map and filter

#map() function returns a map object(which is an iterator) of the results after applying the given function to each item of a given iterable (list, tuple etc.

```python
seq = [1,2,3,4,5]
```

```python
times2
```

```python
map(times2,seq)
```

```python
list(map(times2,seq))
```

```python
list(map(lambda var: var*2,seq))
```

```python
filter(lambda item: item%2 == 0,seq)
```

```python
list(filter(lambda item: item%2 == 0,seq))
```
